**Resveratrol**

```python
pdb_file = '/content/drive/MyDrive/Colab_Notebooks/JC-structural/resv_sirt1.pdb'
ligand_residue = 'RSV'  # Replace 'LIG' with the actual residue name of your ligand
active_residues = [289, 291, 295, 292, 294, 304, 298, 300, 212, 214, 301]
```

```python
def visualize_pdb_with_ligand(pdb_file, ligand_residue, active_residues):
    # Create the Py3Dmol viewer object
    viewer = py3Dmol.view(width=600, height=400)

    # Load the receptor PDB file
    viewer.addModel(open(pdb_file, 'r').read(), 'pdb')

    # Select ligand atoms
    viewer.setStyle({'resn': ligand_residue}, {'stick': {'colorscheme': 'magentaCarbon'}})

    # Set the receptor visualization style
    viewer.setStyle({'chain': 'A'}, {'cartoon': {'color': 'lightgrey'}})

    # Highlight active residues based on residue numbers
    viewer.setStyle({'resi': active_residues}, {'cartoon': {'color': 'lightpink'}})

    # Add labels to ligand residues, f' gives the name
    viewer.addLabel(f'resveratrol', {'fontColor': 'black', 'backgroundColor': 'white', 'fontSize': 10.5}, {'resn': ligand_residue})

    # Set the view to focus on the ligand
    viewer.zoomTo({'chain': 'A', 'resn': ligand_residue})

    # Display the viewer
    viewer.show()

visualize_pdb_with_ligand(pdb_file, ligand_residue, active_residues)
```
